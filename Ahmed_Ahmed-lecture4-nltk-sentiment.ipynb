```python
#Liu and Hu  function 

def liu_hu_lexicon(sentence,plot=True):
    from nltk.corpus import opinion_lexicon
    from nltk.tokenize import treebank
    import pylab
    
    #from list of sentences to one sentence 
    sentence = ''.join([str(tweet) for tweet in sentence])
    
    tokenizer = treebank.TreebankWordTokenizer()
    pos_words = 0
    neg_words = 0
    tokenized_sent = [word.lower() for word in tokenizer.tokenize(sentence)]
    
    x = list(range(len(tokenized_sent)))
    y = []
    
    for word in tokenized_sent:
        if word in opinion_lexicon.positive():
            pos_words += 1
            y.append(1)
        elif word in opinion_lexicon.negative():
            neg_words +=1
            y.append(-1)
        else:
            y.append(0)
            
    
    if pos_words > neg_words: 
        print(str(10*pos_words/len(sentence.split())) +" positive")
    elif pos_words < neg_words:
        print(str(10*neg_words/len(sentence.split())) +"negative")
    elif pos_words == neg_words:
        print("neutral")
    
    if plot == True:
        pylab.plot(x,y)
        pylab.ylabel(["Negative","Neutral","Positive"])
        
```

```python
#function that uses the vader approach 

def vader_function(text):
    from nltk.sentiment import SentimentIntensityAnalyzer
    
    text = ''.join([str(tweet) for tweet in text])

    
    vader_analyzer = SentimentIntensityAnalyzer()
    print(vader_analyzer.polarity_scores(text))
```
